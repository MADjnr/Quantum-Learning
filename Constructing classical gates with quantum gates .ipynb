```python
import qiskit
import numpy as np
from qiskit import Aer, QuantumCircuit, execute, ClassicalRegister, QuantumRegister, BasicAer
from qiskit.visualization import plot_histogram # 2D histogram
from math import pi, sqrt #pi = 3.142 and square root operation
```

```python
#NOT GATE
qc_not = QuantumCircuit(1,1, name = 'qc')
qc_not.h(0)
qc_not.measure(0,0)
qc_not.draw('mpl')
```

```python
creg = ClassicalRegister(1)
Qreg = QuantumRegister(1)
circuit = QuantumCircuit(creg,Qreg)
circuit.draw('mpl')
```

```python
#use the Aer's qasm simulator
backend = Aer.get_backend('qasm_simulator')

#to execute the circuit on the qasm simulator
job = execute(qc_not,backend, shots = 1000)

#to grab the result
result = job.result()

#to return counts
counts = result.get_counts(qc_not)
print('the total counts are', counts)

#plot the histogram
plot_histogram(counts)
```

```python
#FOR THE AND GATE
qc_and = QuantumCircuit(3,1, name = 'qc')
qc_and.x(0)
qc_and.x(1)

qc_and.ccx(0,1,2)
qc_and.measure(2,0)
qc_and.draw('mpl')
```

```python
#to confirm the result in the qasm simulator

backend = Aer.get_backend('qasm_simulator')
job = execute(qc_and, backend, shots = 1000)
results = job.result()
counts = results.get_counts(qc_and)
print('the total counts ', counts)
plot_histogram(counts)
```

```python
#FOR THE OR GATE

qc_or = QuantumCircuit(3,1, name ='qc')
qc_or.x(0)
qc_or.x(1)

qc_or.cx(1,2)
qc_or.cx(0,2)
qc_or.ccx(0,1,2)
qc_or.measure(2,0)
qc_or.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(qc_or, backend, shots = 1000)
results = job.result()
counts = results.get_counts(qc_or)
print('the counts is equal', counts)
plot_histogram(counts)
```

```python
#XOR GATE 
qc_xor = QuantumCircuit(3,1, name = 'qc')
qc_xor.x(0)
qc_xor.cx(1,2)
qc_xor.cx(0,2)
qc_xor.measure(2,0)
qc_xor.draw('mpl')
```

```python
#Half Adder Circuit Implementation


qc_ha = QuantumCircuit(4,2, name = 'qc')

#initialize the qubits to |11>
qc_ha.x(0)
qc_ha.x(1)


#XOR
qc_ha.cx(1,2)
qc_ha.cx(0,2)
qc_ha.barrier()

#AND 
qc_ha.ccx(0,1,3)

#SUM
qc_ha.measure(2,0)

#Carry
qc_ha.measure(3,1)

qc_ha.draw('mpl')
```

```python
backend = Aer.get_backend('qasm_simulator')
job = execute(qc_ha, backend, shots = 1000)
results = job.result()
counts = results.get_counts(qc_ha)
print('the counts for the circuit is ', counts)

plot_histogram(counts)
```
